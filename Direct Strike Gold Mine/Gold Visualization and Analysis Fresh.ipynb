```python
import pandas as pd
import plotly.graph_objects as go
#notes
#run 1 is 0 gms
#run 2 is 1 gms
#run 3 is 2 gms
#run 4 is 3 gms
#run 5 is 4 gms
#run 6 is 1v4
#run 7 is 4v1
#run 8 is 1v3
#run 9 is 3v1
#run 10 is Gms according to guide
#run 11 is Gms according to guide + .4 mid bonus until 12
#run 12 is 3vGuide BAD DONT USE
#run 13 is 4vGuide
#run 14 is 3vGuide
#run 15 is 2vguide

# Load your CSV (if not already loaded)
df = pd.read_csv("gold_progress.csv")

# Convert seconds to minutes:seconds
df['minutes:seconds'] = df['seconds'].apply(lambda x: f"{x // 60}:{x % 60:02}")
df.set_index('minutes:seconds', inplace=True)

# Plot using Plotly
fig = go.Figure()

for i in range(1, 6):
    fig.add_trace(go.Scatter(
        x=df.index,
        y=df[f'run_{i}_gold'],
        mode='lines',
        name=f'Run {i}'
    ))

fig.update_layout(
    title='Gold Accumulation Over Time for Each Run',
    xaxis_title='Time (minutes:seconds)',
    yaxis_title='Gold',
    legend_title='Runs',
    hovermode='x unified'
)

fig.show()
```

```python
import pandas as pd
import requests

# Step 1: Load data
url = "https://website-backend.w3champions.com/api/w3c-stats/games-lengths"
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

# Step 2: Filter to just GM_DS (gameMode == 1101)
df_ds = df[df["gameMode"] == 1101].copy()

# Step 3: Explode the 'lengths' list into individual rows
df_ds = df_ds.explode("lengths").reset_index(drop=True)

# Step 4: Turn the 'lengths' dictionaries into columns
df_ds_expanded = pd.concat([
    df_ds.drop(columns="lengths"),
    df_ds["lengths"].apply(pd.Series)
], axis=1)

# Rename for clarity
df_ds_expanded.rename(columns={"seconds": "game_length_seconds", "games": "num_games"}, inplace=True)

# Step 5: Perform analysis
# Average game length (weighted mean)
average_duration = (df_ds_expanded["game_length_seconds"] * df_ds_expanded["num_games"]).sum() / df_ds_expanded["num_games"].sum()

# Median (approximate based on frequency)
df_sorted = df_ds_expanded.loc[df_ds_expanded["num_games"] > 0].sort_values("game_length_seconds").reset_index(drop=True)
df_sorted["cumulative"] = df_sorted["num_games"].cumsum()
half_total = df_sorted["num_games"].sum() / 2
median_duration = df_sorted[df_sorted["cumulative"] >= half_total]["game_length_seconds"].iloc[0]

# Mode = game length with most games
mode_duration = df_ds_expanded.loc[df_ds_expanded["num_games"].idxmax(), "game_length_seconds"]

# Display results
print(f"ðŸŽ® GM_DS Game Duration Stats:")
print(f"â€¢ Average: {average_duration:.2f} seconds ({average_duration/60:.2f} minutes)")
print(f"â€¢ Median: {median_duration} seconds ({median_duration/60:.2f} minutes)")
print(f"â€¢ Mode: {mode_duration} seconds ({mode_duration/60:.2f} minutes)")

import numpy as np


expanded_durations = np.repeat(
    df_ds_expanded["game_length_seconds"].values,
    df_ds_expanded["num_games"].values
)


percentiles = [5,10,25, 50, 75, 90, 95]
percentile_values = np.percentile(expanded_durations, percentiles)


print("ðŸ“Š GM_DS Game Duration Percentiles:")
for p, v in zip(percentiles, percentile_values):
    print(f"â€¢ {p}th percentile: {v:.0f} seconds ({v/60:.2f} minutes)")
```

```python
import plotly.express as px

# Sort by game length 
df_plot = df_ds_expanded.sort_values("game_length_seconds")

# Convert seconds to minutes
df_plot["minutes"] = df_plot["game_length_seconds"] / 60

# Create interactive bar chart
fig = px.bar(
    df_plot,
    x="minutes",              # X-axis: Game length in minutes
    y="num_games",            # Y-axis: Number of games
    title="GM_DS Game Length Distribution (W3Champions)",
    labels={"minutes": "Game Length (Minutes)", "num_games": "Number of Games"},
    hover_data={"game_length_seconds": True, "num_games": True},
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
